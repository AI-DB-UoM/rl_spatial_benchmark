```python
class TreeProperties:
    def __init__(self, **kwargs):
        
        self.Dimension = kwargs.get('Dimension', None)
        self.FillFactor = kwargs.get('Fill factor', None)
        self.IndexCapacity = kwargs.get('Index capacity', None)
        self.LeafCapacity = kwargs.get('Leaf capacity', None)
        self.TightMbrs = kwargs.get('Tight MBRs', None)
        self.NearMinimumOverlapFactor = kwargs.get('Near minimum overlap factor', None)
        self.ReinsertFactor = kwargs.get('Reinsert factor', None)
        self.SplitDistributionFactor = kwargs.get('Split distribution factor', None)
        self.Utilization = kwargs.get('Utilization', None)
        self.Reads = kwargs.get('Reads', None)
        self.Writes = kwargs.get('Writes', None)
        self.Hits = kwargs.get('Hits', None)
        self.Misses = kwargs.get('Misses', None)
        self.TreeHeight = kwargs.get('Tree height', None)
        self.NumberOfData = kwargs.get('Number of data', None)
        self.NumberOfNodes = kwargs.get('Number of nodes', None)
        self.Splits = kwargs.get('Splits', None)
        self.Adjustments = kwargs.get('Adjustments', None)
        self.QueryResults = kwargs.get('Query results', None)
        self.BufferHits = kwargs.get('Buffer hits', None)
        self.IndexId = kwargs.get('IndexId', None)
        self.Status = kwargs.get('Status', None)
        self.ElapsedTime = kwargs.get('Elapsed Time', None)
        self.ElapsedBuildTime = kwargs.get('Elapsed Learn Time', 0)
        self.ElapsedLearnTime = kwargs.get('Elapsed Build Time', 0)
        if not self.ElapsedTime:
            # if self.ElapsedBuildTime and self.ElapsedLearnTime:
            self.ElapsedTime = self.ElapsedBuildTime + self.ElapsedLearnTime

        self.KnnQuery = kwargs.get('knn query', None)
        self.IndexedSpace = kwargs.get('Indexed space', None)
        self.Operations = kwargs.get('Operations', None)
        self.IndexIo = kwargs.get('Index I/O', None)
        self.LeafIo = kwargs.get('Leaf I/O', None)
        
        self.QueryNum = kwargs.get('Query num', None)
        self.QueryMean = kwargs.get('Query mean', None)
        self.QueryVariance = kwargs.get('Query variance', None)
        self.QueryStdDev = kwargs.get('Query stdDev', None)
        self.QueryP50 = kwargs.get('Query p50', None)
        self.QueryP99 = kwargs.get('Query p99', None)

        self.InsertNum = kwargs.get('Insert num', None)
        self.InsertMean = kwargs.get('Insert mean', None)
        self.InsertVariance = kwargs.get('Insert variance', None)
        self.InsertStdDev = kwargs.get('Insert stdDev', None)
        self.InsertP50 = kwargs.get('Insert p50', None)
        self.InsertP99 = kwargs.get('Insert p99', None)

    def __str__(self):
        attrs = vars(self)
        return '\n'.join(f"{key}: {value}" for key, value in attrs.items())
```

```python
class DataInfo:
    def __init__(self, size, dimensions, distribution, skewness, bounds):
        self.size = size
        self.dimensions = dimensions
        self.distribution = distribution
        self.skewness = skewness
        self.bounds = bounds
    
    def __str__(self):
        return (f"DataInfo(size={self.size}, dimensions={self.dimensions}, "
                f"distribution={self.distribution}, skewness={self.skewness}, "
                f"bounds={self.bounds})")

class QueryInfo:
    def __init__(self, type, size, dimensions, distribution, skewness, query_range=None, k=None, bounds=None):
        self.query_type = type
        self.size = size
        self.dimensions = dimensions
        self.distribution = distribution
        self.skewness = skewness
        self.query_ranges = query_range
        self.ks = k
        self.bounds = bounds
        self.query_strs = []

        if self.query_type == "range":
            for q_range in self.query_ranges:
                query_str = "x".join([str(_) for _ in q_range])
                self.query_strs.append(query_str)

    def __str__(self):
        return (f"QueryInfo(type={self.query_type}, size={self.size}, dimensions={self.dimensions}, "
                f"distribution={self.distribution}, skewness={self.skewness}, "
                f"query_ranges={self.query_ranges}, ks={self.ks}, bounds={self.bounds})")

class BaselineInfo:
    def __init__(self, name, config):
        self.name = name
        self.config = config
    
    def __str__(self):
        return f"BaselineInfo(name={self.name}, config={self.config})"

class Experiment:
    def __init__(self, data_info, query_info, baseline_info):
        self.data_info = data_info
        self.query_info = query_info
        self.range_query_info = list(filter(lambda x: x.query_type == 'range', query_info))
        self.knn_query_info = list(filter(lambda x: x.query_type == 'knn', query_info))
        self.baseline_info = baseline_info

    def __str__(self):
        queries_str = "\n  ".join([str(q) for q in self.query_info])
        baseline_str = "\n  ".join([str(b) for b in self.baseline_info])
        return (f"Experiment:\n"
                f"Data Info:\n  {self.data_info}\n"
                f"Queries Info:\n  {queries_str}\n"
                f"Baseline Info:\n  {baseline_str}")

    def is_baseline(self, baseline_name):
        for info in self.baseline_info:
            if info.name == baseline_name:
                return True, f"result/libspatialindex/{baseline_name}"
        return False, None

    def is_data(self, data_size, dimensions, distribution, skewness):
        if self.data_info.size != data_size:
            return False, None
        # if self.data_info.dimensions != dimensions:
        #     return False
        if self.data_info.distribution != distribution:
            return False, None
        if self.data_info.skewness != skewness:
            return False, None
        return True, f"data_{data_size}_{dimensions}_{distribution}_{skewness}"

    def is_query(self, baseline, type, size, dimensions, distribution, skewness, query_str=None, k=None):
        for info in self.query_info:
            if info.query_type != type:
                continue
            if info.size != size:
                continue
            # if info.dimensions != dimensions:
            #     continue
            if info.distribution != distribution:
                continue
            if info.skewness != skewness:
                continue
            if info.query_type == "range" and query_str not in info.query_strs:
                continue
            if info.query_type == "knn" and k not in info.ks:
                continue
            if type == "range":
                res = f"{type}_{size}_{dimensions}_{distribution}_{skewness}_{query_str}"
            else:
                if baseline == "bmtree":
                    res = f"range_{size}_{dimensions}_{distribution}_{skewness}_{query_str}_depth_1_{type}_{size}_{dimensions}_{distribution}_{skewness}_k_{k}"
                else:
                    res = f"{type}_{size}_{dimensions}_{distribution}_{skewness}_k_{k}"
            return True, res
        return False, None
```

```python
def get_experimental_results(exp_config_file):
    
    experiments = []
    with open(exp_config_file, "r") as exp_config:
        config = json.load(exp_config)

    for index, experiment in enumerate(config["experiments"]):
        
        data_info = DataInfo(**experiment["data"])
        
        query_info = [QueryInfo(**query) for query in experiment["query"]]
        
        baseline_info = [BaselineInfo(name=baseline["name"], config=baseline["config"]) for baseline in experiment["baseline"]]
        
        exp = Experiment(data_info=data_info, query_info=query_info, baseline_info=baseline_info)

        experiments.append(exp)

    return experiments
```

```python
def get_data_name(data_info):
    if data_info.distribution in ['normal', 'skewed', 'uniform']:
        data_file_name = SYNTHETIC_DATA_FILENAME_TEMPLATE.format(
            size=data_info.size,
            dimensions=data_info.dimensions,
            distribution=data_info.distribution,
            skewness=data_info.skewness
        )
    else:
        data_file_name = RELATIVE_REAL_DATA_FILENAME.format(
            data_distribution=data_info.distribution,
            data_size=data_info.size
        )
    return data_file_name.rstrip('.csv')
        
```

```python
def get_range_query_name(data_file_prefix, data_info, query_range, query_type, n_queries, dimensions, distribution, skewness):
    query_range_params = ' '.join([str(val) for val in query_range])
    range_str = "x".join([str(_) for _ in query_range])
    if data_info.distribution in ['normal', 'skewed', 'uniform']:
        file_name = RANGE_QUERY_FILENAME_TEMPLATE.format(
            query_type=query_type,
            n_queries=n_queries,
            dimensions=dimensions,
            distribution=distribution,
            skewness=skewness,
            range_str=range_str
        )
    else:
        file_name = REAL_RANGE_QUERY_FILENAME_TEMPLATE.format(
            data=data_file_prefix,
            query_type=query_type,
            n_queries=n_queries,
            dimensions=dimensions,
            distribution=distribution,
            skewness=skewness,
            range_str=range_str
        )
    return file_name.rstrip('.csv')
    
```

```python
def get_knn_query_name(data_file_prefix, data_info, query):
    res = []

    if data_info.distribution in ['normal', 'skewed', 'uniform']:
        file_name = KNN_QUERY_FILENAME_TEMPLATE.format(
            query_type=query.query_type,
            n_queries=query.size,
            dimensions=query.dimensions,
            distribution=query.distribution,
            skewness=query.skewness
        )
    else:
        file_name = REAL_KNN_QUERY_FILENAME_TEMPLATE.format(
            data=data_file_prefix,
            query_type=query.query_type,
            n_queries=query.size,
            dimensions=query.dimensions,
            distribution=query.distribution,
            skewness=query.skewness
        )
    file_name = file_name.rstrip('.csv')

    for k in query.ks:
        res.append((k, file_name))
        
    return res
```

```python
def parse_rtree_properties(file_path):
    properties = {}
    with open(file_path, 'r') as file:
        for line in file:
            if ": " in line: 
                key, value = line.strip().split(": ", 1)
                
                if value.isdigit():
                    value = int(value)
                elif value.replace('.', '', 1).isdigit():
                    value = float(value)
                properties[key] = value
    return TreeProperties(**properties)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# 设置字体为Times New Roman
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']
rcParams['text.usetex'] = False  # 确保禁用LaTeX

# 示例数据
datasets = ['Dataset1', 'Dataset2', 'Dataset3']  # 数据集名称
baseline_names = ['Baseline1', 'Baseline2', 'Baseline3', 'Baseline4', 'Baseline5', 'Baseline6', 'Baseline7', 'Baseline8', 'Baseline9']
query_times = np.random.rand(9, len(datasets)) * 100  # 随机生成查询时间

# 定义颜色和填充样式
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0','#FFA07A', '#B0E0E6', '#FFD700', '#D3D3D3']
patterns = ['/', 'x', 'o', 
            '\\', '+', '.', 
            '|', '-', '*']

def plot_hist(datasets, baseline_names, result, y_label="", is_log=False):

    label_size=16
    legend_size=16
    
    # 绘制直方图
    fig, ax = plt.subplots(figsize=(10, 6))
    width = 0.8 / len(baseline_names)  # 条形的宽度
    x = np.arange(len(datasets))
    
    for i, baseline in enumerate(baseline_names):
        ax.bar(x - int(len(baseline_names) / 2) * width + i * width, result[i], width=width, label=baseline, color=colors[i % len(colors)], hatch=patterns[i % len(patterns)])

    # ax.set_xlabel('Dataset Name')
    ax.set_ylabel(y_label, fontsize=label_size)
    # ax.set_title('Query Time by Dataset and Baseline')
    ax.set_xticks(x)
    ax.set_xticklabels(datasets, fontsize=label_size)
    ax.tick_params(axis='x', labelsize=label_size)  # 设置y轴刻度大小
    ax.tick_params(axis='y', labelsize=label_size)  # 设置y轴刻度大小
    if is_log:
        ax.set_yscale('log')
    
    # 将图例放置在图形外侧的右边
    legend = ax.legend(title='Baseline', loc='upper left', bbox_to_anchor=(1,1), fontsize=legend_size)
    plt.setp(legend.get_title(), fontsize=legend_size)  # 设置图例标题的字体大小
    
    plt.tight_layout()  # 自动调整子图参数, 使之填充整个图像区域
    plt.show()
```

```python
def get_exp_data(baseline, type, data_size, data_distribution, data_skewness, query_size, query_distribution, query_skewness, query_str="0.01x0.01", k=25):
    for experiment in experiments:
        res, path = experiment.is_baseline(baseline)
        if not res:
            continue
        res, data = experiment.is_data(data_size, 2, data_distribution, query_skewness)
        if not res:
            continue
        res, query = experiment.is_query(baseline, type, query_size, 2, query_distribution, query_skewness, query_str, k)
        if not res:
            continue
        
        return path, data, query
```

```python
import matplotlib.pyplot as plt

def plot_line(data_map, varying_cardinality, attribute, title, xlabel, ylabel, scale):
    
    plt.figure(figsize=(5, 3))
    for baseline, properties_list in data_map.items():
        # extract passed attribute
        attribute_values = [getattr(properties, attribute, None) / scale for properties in properties_list]
        plt.plot(varying_cardinality, attribute_values, label=baseline, marker='o')

        print(varying_cardinality)
        print(attribute_values)
        
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.yscale('log')
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_hist(data_map, varying_cardinality, attribute, title, xlabel, ylabel, scale):
    
    plt.figure(figsize=(5, 3))
    width = 0.25  # width of the bars
    n = len(data_map)  # number of data groups
    _x = range(len(varying_cardinality))
    
    for i, (baseline, properties_list) in enumerate(data_map.items()):
        # extract passed attribute
        attribute_values = [getattr(properties, attribute, None) / scale for properties in properties_list]
        # Shift bars for each dataset
        plt.bar([x + i * width for x in _x], attribute_values, width=width, label=baseline)

        print(varying_cardinality)
        print(attribute_values)
        
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    # Adjust x-ticks to be in the middle of the grouped bars
    plt.xticks([x + width * (n / 2 - 0.5) for x in _x], varying_cardinality)
    plt.yscale('log')
    plt.legend()
    plt.grid(True)
    plt.show()
```
